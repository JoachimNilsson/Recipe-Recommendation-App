# Recommender system
For my intended use I will need a recommender system that can recommend recipes to a user based on one or more recipes that we know the user likes. My approach will be to calculate similarity scores between recipes, since I don't have data of user-recipe interactions. I will do this be creating a vector space representation of the recipes and calculate the pariwise cosine similarity between the vectors. We can then compute a rating for a recipe with the weighted sum of recipes we know that the user likes $$\text{rating}(I_i) = \frac{\sum_j \text{rating}(I_j)*s_{ij}}{\sum_j s_{ij}}$$ where $s_{ij}$ is the similarity score bwtween recipe $i$ and $j$. Note that in this case the rating is binary, either the user likes the recipe or not. From this we can recommend recipe to the user by ranking recipes by this rating.

## Representing the recipes by their ingredients

### Bag-of-words representation


```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def bag_of_words_matrix(data):

    
    count_vectorizer = CountVectorizer(analyzer=lambda x: x)
    count_matrix = count_vectorizer.fit_transform(data).toarray()
 
    return count_matrix, count_vectorizer
    
def recommend_recipes(urls, recipes, cosine_matrix, n=10):
    indexes = list(recipes[recipes["url"].isin(urls)].index.values)
    print(indexes)
    cosine_matrix = np.delete(cosine_matrix, indexes, 1)
    cosine_matrix_rows = np.take(cosine_matrix, indexes, 0)
    recommend_recipes_indexes = np.mean(cosine_matrix_rows, axis=0).argsort()[-n:][::-1]
    return recipes.iloc[recommend_recipes_indexes]["url"].tolist()

```

```python
count_matrix, count_vectorizer = bag_of_words_matrix(recipes["clean_ingredients"].tolist())
cosine_matrix_count = cosine_similarity(count_matrix)
np.save('cosine_matrix_bow.npy', cosine_matrix_count)
```

## Word2Vec
The main limitation of the two previous representations is that they do not capture how ingredients co-occur with other ingredients. By using the word2vec-model by gensim we can capture similarity between ingredients in regard to the context they occur in and thereby also similarity between recipes. As we do not care about the order of the ingredients in a recipe we choose to sort the ingredients alphabetical order.

```python
recipes["sorted_clean_ingredients"] = recipes["clean_ingredients"].apply(lambda x: sorted(x))
recipes.head()
```

```python
max([len(recipe) for recipe in recipes["sorted_clean_ingredients"].tolist()])
```

```python
from gensim.models import Word2Vec

num_features = 120    # Word vector dimensionality                      
min_word_count = 1                        
num_workers = 4       # Number of CPUs
context = 100         # Context window size; 
sg = 0                # continuous bag of words
downsampling = 1e-2   # threshold for configuring which 
                      # higher-frequency words are randomly downsampled

model = Word2Vec(recipes["sorted_clean_ingredients"].tolist() , workers=num_workers, \
            vector_size=num_features, min_count = min_word_count, \
            window = context, sg=sg)

model.save("word2vec.model")
```

```python
from gensim.models import Word2Vec
model = Word2Vec.load("word2vec.model")
model.wv['vetemjöl']
```

```python
import numpy as np
def create_recipe_embedding(ingredients):
    return np.mean(np.array([model.wv[ingredient] for ingredient in ingredients]),0)
len(create_recipe_embedding(recipes["sorted_clean_ingredients"].tolist()[0]))
```

I'll begin by calculating the embedding for a recipe by averaging the ingredient-embeddings for the recipe. I can then calculate the cosine similarity between these recipe-embeddings.

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
recipes["vord_vector"] = recipes["sorted_clean_ingredients"].apply(create_recipe_embedding)
recipes_word2vec = np.vstack(recipes["vord_vector"].tolist())
cosine_matrix_word2vec = cosine_similarity(recipes_word2vec)
np.save('cosine_matrix_word2vec.npy', cosine_matrix_word2vec)

```

```python
cosine_matrix_word2vec = np.load('cosine_matrix_word2vec.npy')
```

```python
import time

def recommend_recipes_word2vec(url, recipes, cosine_matrix_word2vec, n=10):
    ind = recipes.index[recipes['url']==url].tolist()[0]
    recommend_recipes_indexes = cosine_matrix_word2vec[ind].argsort()[-n:][::-1]
    return recipes.iloc[recommend_recipes_indexes]["url"].tolist()[1:]


random_indexes = np.random.choice(np.arange(len(recipes)), 1)
random_urls = recipes["url"].iloc[random_indexes].tolist()

start_time = time.time()
recommend_recipes_word2vec_list = recommend_recipes_word2vec(random_urls[0], recipes, cosine_matrix_word2vec, n=10)
print("--- %s seconds ---" % (time.time() - start_time))

print(*random_urls, sep="\n")
print("Similar recipes word2vec:", *recommend_recipes_word2vec_list, sep="\n")

```

The recommendations when having more than one liked recipe does not make much sense.
